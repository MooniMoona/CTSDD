This is a visualization code for wind turbine monitoring data restoration and wind turbine blade icing warning. After performing the detection task (`exe_ice_warning.py`), it can be plotted based on the results obtained.

We have given example results (`./save/test_results_examples/...`). 

If you want to visualize the new detection results, you can replace the path in the code with your result path (`./save/test_results/...`).

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import torch
import pickle
import os
import pickle
from plot_results import get_results_quantile, wp_quantiles_plot, plot_case_scores
from matplotlib import rcParams

rcParams['font.family'] = 'Times New Roman'
rcParams['font.size'] = 12

config = 'WT_ice'
model = 'CTSDD'
nsample = 5 # 5 runs per detection
targetstrategy = 'random_mid_wp'
pred_length = 30
random_mask_percent = 0.1
WP_number = 2
mean_std_path = "./dataset/data_mean_std.pk"
threshold_path = './dataset/normal_Robust_threshold.csv'
```

## Normal results

```python
# all restoration data
foldername = './save/test_results_examples/'
select_case_name = 'normal'
save_foldername = foldername + select_case_name+'_test/'
all_target_np, all_evalpoint_np, all_given_np, quantiles_pred \
        = get_results_quantile(nsample, save_foldername, mean_std_path)
        
# obtain WP restoration data
wp_target_np, wp_evalpoint_np, wp_given_np = \
  all_target_np[:, :, WP_number], all_evalpoint_np[:, :, WP_number], all_given_np[:, :, WP_number]
wp_quantiles_np = [wp[:, :, WP_number] for wp in quantiles_pred]
```

### Normal WP restoration results

```python
# choose a sample : dataind
foldername = './save/test_results_examples/'
select_case_name = 'normal'
save_foldername = foldername + select_case_name+'_test/'
dataind = 20
wp_quantiles_plot(wp_target_np, wp_evalpoint_np, wp_given_np, wp_quantiles_np, 
                         dataind,select_case_name,save_foldername,save=1)
```

```python
model_name = 'CTSDD'
foldername = './save/test_results_examples/'
select_case_name = 'normal'
save_foldername = foldername + select_case_name+'_test/'

scores_name = 'combined_mae_dtw_scores'
fault_time = None
threshold_name = 'mae_dtw_threshold'
detect_interval = 5 # Simulate Detection Interval
    
plot_case_scores(model_name, select_case_name, scores_name, save_foldername,
                          threshold_path, fault_time, threshold_name, detect_interval)    
    
```
